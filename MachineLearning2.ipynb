```python
import pandas as pd
```

```python
data = pd.read_csv('Features_Matrix_Labeled.csv')
data = data.dropna()
data
```

```python
features = [col for col in data.columns if col not in ['ID', 'PAINTING', 'voto', 'binary']]
X = data[features]
y = data['voto']
yb = data['binary']
```

```python
import scipy.stats as stats

# Dizionario per conservare i risultati
results = {}

# Calcola correlazione e p-value per ogni colonna in X rispetto a y
for column in X.columns:
    corr, p_value = stats.pearsonr(X[column], y)
    results[column] = {'Pearson Correlation': corr, 'P-value': p_value}

# Converte i risultati in DataFrame per una migliore visualizzazione
results_df = pd.DataFrame(results).T  # Trasponi per una migliore formattazione
print("Risultati Correlazione Pearson e P-value:")
print(results_df)

# Calcola anche per Spearman
results_spearman = {}
for column in X.columns:
    corr, p_value = stats.spearmanr(X[column], y)
    results_spearman[column] = {'Spearman Correlation': corr, 'P-value': p_value}

# Converte i risultati in DataFrame per una migliore visualizzazione
results_spearman_df = pd.DataFrame(results_spearman).T
print("Risultati Correlazione Spearman e P-value:")
print(results_spearman_df)
```

```python
# Kruskal-Wallis Test
kruskal_results = {}
for column in X.columns:
    try:
        group1 = X[column][yb == 1]
        group2 = X[column][yb == 0]
        stat, p_value = stats.kruskal(group1, group2)
        kruskal_results[column] = {'Kruskal-Wallis Stat': stat, 'P-value': p_value}
    except:
        print(f'Failed column {column} in KS')

# T-Test (Assumendo varianze uguali)
ttest_results = {}
for column in X.columns:
    try:
        group1 = X[column][yb == 1]
        group2 = X[column][yb == 0]
        stat, p_value = stats.ttest_ind(group1, group2, equal_var=True)  # `equal_var=False` per Welch's t-test
        ttest_results[column] = {'T-Test Stat': stat, 'P-value': p_value}
    except:
        print(f'Failed column {column} in TTest')

# Mostra i risultati
kruskal_df = pd.DataFrame(kruskal_results).T
ttest_df = pd.DataFrame(ttest_results).T

print("Risultati del Test di Kruskal-Wallis:")
print(kruskal_df)
print("\nRisultati del T-Test:")
print(ttest_df)
```
