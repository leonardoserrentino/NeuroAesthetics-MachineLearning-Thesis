```python
import numpy as np

class_counts = data['Binary'].value_counts()
class_majority = class_counts.idxmax()
class_minority = class_counts.idxmin()

# Filtra le classi
class_majority_data = data[data['Binary'] == class_majority]
class_minority_data = data[data['Binary'] == class_minority]

# Calcola il numero di istanze nella classe minoritaria
n_minority = class_minority_data.shape[0]

# Verifica che il numero di istanze della classe maggioritaria sia sufficiente
if n_minority > class_majority_data.shape[0]:
    raise ValueError("Non ci sono abbastanza istanze nella classe maggioritaria per bilanciare il dataset.")

# Campionamento casuale delle istanze dalla classe maggioritaria
selected_class_majority_data = class_majority_data.sample(n=n_minority, random_state=42)  # Usa random_state per la riproducibilitÃ 

# Crea il dataset bilanciato combinando le due classi
balanced_data = pd.concat([class_minority_data, selected_class_majority_data]).sample(frac=1, random_state=42)  # Mescola le righe

# Identifica gli indici delle istanze della classe maggioritaria non selezionate
unselected_indices = class_majority_data.index.difference(selected_class_majority_data.index)

# Crea il dataset con le istanze non selezionate
discarded_data = class_majority_data.loc[unselected_indices]
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from keras.models import Sequential
from keras.layers import Dense
from joblib import dump

# Preparazione dei dati
X = balanced_data[selected_features]
y = balanced_data['Binary']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scaling dei dati
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Definizione dei modelli
models = {
    "Logistic Regression": LogisticRegression(),
    "SVM (linear kernel)": SVC(kernel='linear'),
    "SVM (poly kernel)": SVC(kernel='poly'),
    "SVM (rbf kernel)": SVC(kernel='rbf'),
    "SVM (sigmoid kernel)": SVC(kernel='sigmoid'),
    "Random Forest": RandomForestClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
}

# Neural Network setup
nn = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dense(64, activation='relu'),
    Dense(2, activation='softmax')
])
nn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Training dei modelli e salvataggio
results = {}
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    results[name] = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred)
    }
    # Salvataggio del modello
    dump(model, f'models/{name}.joblib')

# Training e testing per la rete neurale
nn.fit(X_train_scaled, y_train, epochs=10, batch_size=10, verbose=0)
y_pred_nn = np.argmax(nn.predict(X_test_scaled), axis=-1)
results['Neural Network'] = {
    "Accuracy": accuracy_score(y_test, y_pred_nn),
    "Precision": precision_score(y_test, y_pred_nn),
    "Recall": recall_score(y_test, y_pred_nn)
}
# Salvataggio della rete neurale
nn.save('models/Neural_Network.h5')

# Creazione della tabella dei risultati
results_data = pd.DataFrame(results).transpose()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score
from joblib import load
from keras.models import load_model
import os

# Preparazione dei dati
X_discarded = discarded_data[features]
y_discarded = discarded_data['Binary']

# Scaling dei dati (assumendo che il scaling sia necessario)
scaler = StandardScaler()
X_discarded_scaled = scaler.fit_transform(X_discarded)

# Percorso della cartella che contiene i modelli salvati
models_path = './models'
model_files = [f for f in os.listdir(models_path) if f.endswith('.joblib') or f.endswith('.h5')]

# Creazione di un DataFrame vuoto per i risultati
results = pd.DataFrame()

# Caricamento e testing di ciascun modello
for file in model_files:
    model_path = os.path.join(models_path, file)
    if file.endswith('.joblib'):
        model = load(model_path)
        y_pred = model.predict(X_discarded_scaled)
    elif file.endswith('.h5'):
        model = load_model(model_path)
        y_pred = np.argmax(model.predict(X_discarded_scaled), axis=-1)
    
    # Calcolo delle metriche
    acc = accuracy_score(y_discarded, y_pred)
    prec = precision_score(y_discarded, y_pred)
    rec = recall_score(y_discarded, y_pred)

    # Creazione di un DataFrame temporaneo con i risultati
    temp_data = pd.DataFrame({
        'Model': [file.split('.')[0]],
        'Accuracy': [acc],
        'Precision': [prec],
        'Recall': [rec]
    })

    # Aggiunta del DataFrame temporaneo al DataFrame dei risultati
    results = pd.concat([results, temp_data], ignore_index=True)

# Visualizzazione dei risultati
print(results)
```

2. Tenfold Cross-Validation

```python
import pandas as pd
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

# Preparazione dei dati
X = balanced_data[features]
y = balanced_data['Binary']

# Definizione di un oggetto StratifiedKFold per la cross-validation
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Scorers personalizzati
accuracy_scorer = make_scorer(accuracy_score)
precision_scorer = make_scorer(precision_score)
recall_scorer = make_scorer(recall_score)

# Definizione dei modelli
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "SVM (linear kernel)": SVC(kernel='linear'),
    "SVM (rbf kernel)": SVC(kernel='rbf'),
    "SVM (poly kernel)": SVC(kernel='poly'),
    "SVM (sigmoid kernel)": SVC(kernel='poly'),
    "Random Forest": RandomForestClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier()
}

# Creazione di un DataFrame per i risultati
results = pd.DataFrame(index=models.keys(), columns=['Accuracy', 'Precision', 'Recall'])

# Training e testing dei modelli usando la cross-validation
for name, model in models.items():
    pipeline = make_pipeline(StandardScaler(), model)
    accuracy = cross_val_score(pipeline, X, y, cv=cv, scoring=accuracy_scorer).mean()
    precision = cross_val_score(pipeline, X, y, cv=cv, scoring=precision_scorer).mean()
    recall = cross_val_score(pipeline, X, y, cv=cv, scoring=recall_scorer).mean()
    results.loc[name] = [accuracy, precision, recall]

# Visualizzazione dei risultati
print(results)
```
